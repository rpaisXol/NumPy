```python
import numpy as np
```

#### 4-1 1차원 배열의 정렬

```python
arr = np.random.randint(10, size=10)
print(arr)
```

```python
print(np.sort(arr)) # 정렬
print(np.sort(arr)[::-1]) # 역순 정렬
```

```python
arr = np.sort(arr)
print(arr)
print('\n')

arr.sort()
print(arr)
```

#### 4-2 2차원 배열의 정렬

```python
arr = np.random.randint(15, size=(3,4))
print(arr)
```

```python
print(np.sort(arr, axis=0)) # 원소가 오름차순 정렬
print('\n')

print(np.sort(arr, axis=None)) # 2차원 -> 1차원 변경 후 오름차순 정렬
```

```python
# argsort() : 정렬된 원소의 원래 배열의 인덱스를 반환한다.
print(np.sort(arr, axis=1))
print('\n')

print(np.argsort(arr, axis=1)) # 행을 기준으로 원래 인덱스를 반환한다. 
```

```python
# argsort() : 정렬된 원소의 원래 배열의 인덱스를 반환한다.
print(np.sort(arr, axis=0))
print('\n')

print(np.argsort(arr, axis=0)) # 열을 기준으로 원래 인덱스를 반환한다. 
```
