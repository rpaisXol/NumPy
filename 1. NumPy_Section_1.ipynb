#### 1-1. N차원 배열 생성하기 

```python
import numpy as np
```

```python
# 1차원 배열
arr = np.array([1,2,3])
print(arr)
```

```python
type([1,2,3])
type(arr)
```

```python
# 2차원 배열
arr = np.array([[1,2,3],
               [4,5,6]])
print(arr)
```

```python
# 1차원 배열
tpl = (4,5,6)
arr= np.array(tpl)
print(arr)
```

```python
# 1차원 배열
lst = [1,2,3]
arr = np.array(lst)
print(arr)
```

```python
# 2차원 배열
lst = [[1,2,3],[4,5,6]]
arr = np.array(lst)
print(arr)
```

```python
# shape : 배열의 크기
arr1 = np.array([1,2,3])
arr2 = np.array([[1,2,3],[4,5,6]])

print(arr1.shape, arr2.shape)
```

```python
# size
print(arr1.size, arr2.size) # 원소 3개, 원소 6개
```

#### 1-2. N차원 배열의 데이터 타입

```python
# 실수형 데이터 타입
arr = np.array([1,2,3], dtype=np.float)
print(arr, arr.dtype)
```

```python
# 정수형 데이터 타입
arr = np.array([1.1,2.2,3.3], dtype=np.int)
print(arr, arr.dtype)
```

```python
# boolean형 데이터 타입
arr = np.array([0,1,1], dtype=np.bool)
print(arr, arr.dtype)
```

```python
# 정수형
arr = np.array([0,1,2,3])
print(arr, arr.dtype)

# 실수형
arr = arr.astype(np.float32)
print(arr, arr.dtype)
```

```python
# 데이터 타입이 혼재하는 경우(정수 + 실수)
arr = np.array([1, 2, 3.4])
print(arr, arr.dtype)
```

```python
# 데이터 타입이 혼재하는 경우(정수 + 실수 + 문자열)
arr = np.array([1, 2, 3.4, "32"])
print(arr, arr.dtype) # unicode로 변경
```

```python
arr = np.array([1, 2, 3.4, "32"], dtype=int)
print(arr, arr.dtype) # 정수형 변경
```

```python
arr = np.array([1, 2, 3.4, "32", "String"], dtype=int)
print(arr, arr.dtype) # Error
```

#### 1-3 정해진 형식의 N차원 배열 생성하기

```python
# np.zeros() : 모든 원소들이 0으로 된 배열을 생성한다.
arr = np.zeros([2,2])
print(arr)
```

```python
# np.ones() : 모든 원소들이 1로 된 배열을 생성한다.
arr = np.ones([3,5])
print(arr)
```

```python
# np.full() : 모든 원소들이 지정된 full-value로 된 배열을 생성한다.
arr = np.full((2,3), 5) # 2 by 3의 원소 값 5
print(arr)
```

```python
# np.eye() : n by m 행렬일때, 대각 원소가 1인 행렬을 생성한다.
# k : 대각원소를 발생할 지점을 선택한다.
arr1 = np.eye(3, 4, k=0) 
print(arr1)
print('\n')

arr2 = np.eye(3, 4, k=1)
print(arr2)
print('\n')

arr3 = np.eye(3, 4, k=-1)
print(arr3)
print('\n')

arr4 = np.eye(3) # 정방 행렬
print(arr4)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

# np.zeros_like() : 원래의 배열과 동일한 크기를 유지한 상태로 원소를 0으로 지정한 배열 생성
arr_z = np.zeros_like(arr)
print(arr_z)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

# np.ones_like() : 원래의 배열과 동일한 크기를 유지한 상태로 원소를 1으로 지정한 배열 생성
arr_o = np.ones_like(arr)
print(arr_o)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

# np.full_like() : 원래의 배열과 동일한 크기를 유지한체 원소를 0으로 지정한 배열 생성
arr_f = np.full_like(arr, 9) # fill_value를 지정해줘야 한다.
print(arr_f)
```

#### 1-4. 특정 범위의 값을 가지는 N차원 배열 생성하기

```python
lst = list(range(0,9,2))
print(lst)
```

```python
arr = np.arange(9) # 0~8
print(arr)
```

```python
arr = np.arange(3, 12) # start
print(arr)
```

```python
arr = np.arange(3, 13, 3) # start, end, step
print(arr)
```

```python
arr = np.arange(start=3, stop=13, step=3) # start, end, step
print(arr)
```

```python
# np.linspace() : 설정한 범위에서 선형적으로 분할한 위치의 값을 출력한다.

arr = np.linspace(0, 100, 11) # 0부터 100사이에 11개의 원소를 균등한 간격으로 배열을 생성한다.
print(arr)
```

```python
arr = np.linspace(0, 100, 100) # 0부터 100사이에 100개의 원소를 균등한 간격으로 배열을 생성한다.
print(arr)
```

```python
# np.logspace() : 설정한 범위에서 로그로 분할한 위치의 값을 출력한다.

arr = np.linspace(1, 10, 10)
print(arr, end='\n\n')

arr = np.logspace(1, 10, 10, base=2) ## 1 ~ 10까지 10개의 원소에서 log값 출력  base 옵션은 log의 밑이다. 
print(arr) # 밑이 2인 log 출력
```

```python
arr = np.logspace(1, 10, 10) # base의 default는 상용로그 log = 10 이다.
print(arr)
```

#### 1-5. 난수로 N차원 배열 생성하기

```python
import matplotlib.pyplot as plt
```

```python
# np.random.normal() : np의 random 생성

arr = np.random.normal(0, 1, 10) # (정규분포 평균, 표준편차, 추출할 표본의 갯수)
print(arr)
```

```python
arr = np.random.normal(0, 1, (2,3)) # (정규분포 평균, 표준편차, 추출할 표본의 갯수(2 by 3))
print(arr) 
```

```python
arr = np.random.normal(0, 1, 100000)
plt.hist(arr, bins=100) # bins : 난수의 갯수를 몇개로 구분해서 보여줄지 지정한다.
plt.show()
```

```python
# np.random.rand() : 0~1 사의 값을 균등한 비율로 표본을 추출한다.
arr = np.random.rand(1000) # 표본의 갯수
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randn() : -1~1 사이(가우시안 분포)의 표본을 추출한다.
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randint() : random한 정수형 int만 출력하는 함수.
arr = np.random.randint(low=1, high=5, size=10) # 1~5까지 size의 모양대로 배열을 생성한다.
print(arr)
```

```python
arr = np.random.randint(low=1, high=5, size=(3,4)) # 1~5까지 size(3 by 4)의 모양대로 배열을 생성한다.
print(arr)
```

```python
arr = np.random.randint(5) # 0~5 사이의 하나의 정수만 반환
print(arr)
```

```python
arr = np.random.randint(100, 200, 1000)
plt.hist(arr, bins=100)
plt.show()
```

#### 1-6. Seed값을 통한 난수 생성 제어

```python
arr1 = np.random.rand(10)
print("난수 발생1 \n", arr1)
print('\n')

arr2 = np.random.rand(10)
print("난수 발생2 \n", arr2)
```

```python
# np.random.seed() : seed를 통해 난수의 생성 패턴을 동일하게 관리한다.
np.random.seed(10)
arr1 = np.random.rand(10)
print("난수 발생1 \n", arr1)
print('\n')

arr2 = np.random.rand(10)
print("난수 발생2 \n", arr2)
```
