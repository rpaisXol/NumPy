#### 6-1. 배열에 원소 추가 및 삭제

```python
arr = [1,2,3,4,5,6,7,8]
arr.insert(2, 50)
print(arr)
```

```python
# 1차원 배열
arr = np.arange(1, 9)
arr = np.insert(arr, 2, 50)
print(arr)
```

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
arr = np.insert(arr, 2, 50, axis=0) # 행 기준 삽입
print(arr)
```

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
arr = np.insert(arr, 2, 50, axis=1) # 행 기준 삽입
print(arr)
```

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
arr = np.insert(arr, 2, 50, axis=None) # 1차원으로 변경 후 삽입 
print(arr)
```

```python
# delete()
arr = np.arange(1, 13).reshape(3, 4)
print(arr)
print('\n')

arr = np.delete(arr, 2, axis=0) # 행 기준 삭제
print(arr)
print('\n')

arr = np.delete(arr, 2, axis=1) # 행 기준 삭제
print(arr)
print('\n')

arr = np.delete(arr, 2, axis=None) # 1차원으로 변경 후 삭제
print(arr)
print('\n')
```

#### 6-2 배열 간의 병합

```python
arr1 = np.arange(1, 13).reshape(3, 4)
arr2 = np.arange(13, 25).reshape(3, 4)

print(arr1)
print('\n')
print(arr2)
print('\n')

# axis = 0
arr3 = np.append(arr1, arr2, axis=0) # 배열 병합(행 기준)
print(arr3, end='\n\n')

# axis = 1
arr3 = np.append(arr1, arr2, axis=1) # 배열 병합(열 기준)
print(arr3, end='\n\n')

# axis = None
arr3 = np.append(arr1, arr2, axis=None) # 배열 병합(1차원 변경 후 기준)
print(arr3, end='\n\n')
```

```python
# vstack(), hstack()
arr1 = np.arange(1, 7).reshape(2, 3)
arr2 = np.arange(7, 13).reshape(2, 3)

arr3 = np.vstack((arr1, arr2)) # 2차원 배열을 axis=0을 기준으로 병합
print(arr3)
print('\n')

arr3 = np.hstack((arr1, arr2)) # 2차원 배열을 axis=1을 기준으로 병합
print(arr3)
```

```python
# concatenate() : axis 값을 기준으로 두 행렬을 병합하는 함수이다.
arr1 = np.arange(1, 7).reshape(2, 3)
arr2 = np.arange(7, 13).reshape(2, 3)

arr3 = np.concatenate([arr1, arr2], axis=0)
print(arr3)
print('\n')

arr3 = np.concatenate([arr1, arr2], axis=1)
print(arr3)
```

#### 6-3 배열의 분할

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
print(arr)
print('\n')

# vsplit() : axis=0을 기준으로 배열을 분할
arr_vsplit = np.vsplit(arr, 3)
print(arr_vsplit)
print('\n')

# hsplit() : axis=1을 기준으로 배열을 분할
arr_hsplit = np.hsplit(arr, 2)
print(arr_hsplit)
```

```python
# 3차원 배열
arr = np.random.randint(0, 10, (4, 6, 8))
print(arr)
```

```python
arr_vsplit = np.vsplit(arr, 2)
print(arr_vsplit)
```

```python
arr_hsplit = np.hsplit(arr, 2)
print(arr_hsplit)
```
